```python
for each in states_ref:
    unemployment_df = unemployment_df.replace([each], states_ref[each])

def clean(x):
    return x[:-1]

unemployment_df['county'] = unemployment_df['county'].apply(clean)

unemployment_df
```

Merge with collapsed county data

```python
import numpy as np
hashmap = {}

for index, row in collapsed_county_df.iterrows():
    combined_key = row['projectcountyname'] + row['projectstate']
    hashmap[combined_key] = row['currentapprovalamount']

    

count = 0
unemployment_df = unemployment_df.assign(ppp=np.nan)
for index, row in unemployment_df.iterrows():
    combined_key = row['county'] + row['countytitle2']
    if combined_key in hashmap:
        count+=1
        ppp_amount = hashmap[combined_key]
        unemployment_df.at[index, 'ppp'] = ppp_amount
    
print(count)    
collapsed_county_df
unemployment_df
```

drop NA

```python
unemployment_df.dropna(subset=['ppp'])
unemployment_df
```
